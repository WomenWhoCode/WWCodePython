```python
import pandas as pd
pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 1000)

csv_file = './world_pop.csv'
df = pd.read_csv(csv_file)
```

```python
# check general information of the dataframe
def basic_info(df,cmd, num:int=0):
  if 'column' in cmd:
    print('list of dataframe columns: {}'.format(df.columns))
  elif 'head' in cmd:
    print('first {} rows of the dataframe:\n {}'.format(num, df.head(num)))
  elif 'tail' in cmd:
    print('last {} rows of the dataframe:\n {}'.format(num, df.tail(num)))
  elif 'every' in cmd:
    print('every {} row of the dataframe:\n {}'.format(num, df.iloc[::num, :]))
  elif 'shape' in cmd:
    print('shape of the dataframe: {}'.format(df.shape))
  elif 'size' in cmd:
    print('number of rows in the dataframe: {}'.format(df.size))
  else:
    pass
```

```python
# get all the unique values of 'Location' columns
print(list(df['Location'].unique()))
```
