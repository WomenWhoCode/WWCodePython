## datetime module

```python
import datetime  # we don't need to install it separately
```

### Functions

```python
print(datetime.datetime.now()) # returns current date and time in UTC
```

```python
print(datetime.date.today()) # returns today's date
```

```python
print(datetime.date(2021,6,12)) # returns YYYY-MM-DD
```

```python
print(datetime.date.fromtimestamp(1627481187)) # returns YYYY-MM-DD from timestamp (Unix epoch time)
```

```python
print(datetime.time(2,3,2,2)) # returns HH:MM:SS.MS
```

Write a Python script to display the various Date Time formats

a) Current date and time

b) Current year

c) Month of year

d) Week number of the year

image.png

```python
import datetime
print("Current date and time: " , datetime.datetime.now())
print("Current year: ", datetime.date.today().strftime("%Y"))
print("Month of year: ", datetime.date.today().strftime("%B"))
print("Week number of the year: ", datetime.date.today().strftime("%W"))
```

## time module

```python
import time
```

### Functions

```python
print(time.time()) # returns the number of seconds passed since epoch 
#(epoch refers to poch is the point where the time starts, and is platform dependent. For Unix, the epoch is January 1, 1970, 00:00:00 (UTC))
```

```python
time.ctime(1) # takes seconds passed since epoch as an argument and returns a string representing local time
```

```python
time.sleep(12) # delays execution of the current thread for the given number of seconds
```

```python
print("Printed immediately.")
time.sleep(2.4)
print("Printed after 2.4 seconds.")
```
