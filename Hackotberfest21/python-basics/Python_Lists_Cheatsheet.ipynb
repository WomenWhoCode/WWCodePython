## **Lists in Python**
Lists are built-in data structures in Python that can contain an ordered set of elements. This is a mutable data structure, which means that once the list is created, it can be changed by either appending, replacing or removing elements.
The values in a list, need not be unique. 

> **These are typically identified as a set of elements, surrounded by [ ]. The elements inside the list are seperated by commas.**

### **How to create a List?**
Lists can be created in two ways:

1.   Using [ ] directly
2.   Using the list( ) constructor - This can be used to create lists from other sequences and iterators (such as strings and dictionaries)


```python
#Using []
l=["Python","Java","C","C#","JavaScript"]
num_list=[3,4,2,2,1,1,5,6]
print(l)
print(num_list)
```

```python
#Using list()

#From a tuple
t_list=list(("Python","Java","C","C#","JavaScript"))
print("List from a tuple:",t_list) #['Python', 'Java', 'C', 'C#', 'JavaScript']

#From a string
s="Python"
s_list=list(s)
print("List from a string:",s_list) #['P', 'y', 't', 'h', 'o', 'n']

#From a dictionary
d={"A":1,"B":4,"C":3}
d_list=list(d)
print("List from a dictionary:",d_list) #['A', 'B', 'C']
```

### **Indexing & Slicing Lists**


*   Python follows zero-based indexing
*   Negative indices are used to access elements from the end of the list
*   The **:** [colon] operator is used for slicing. 
*   The format for slicing is **[start:stop:step]**







```python
#Accessing an element
print("The list is:",l)
print("The first element is:",l[0]) #Python
print("The second element is:",l[1]) #Java
print("The final element is:",l[-1]) #JavaScript
l[0]="Java"
print("The list after replacing one element with another is:",l) #['Java', 'Java', 'C', 'C#', 'JavaScript']

l[0]="Python" #changing it back to its original value
```

```python
#Slicing [start:stop:step]
#Use :: only when the step has to be accessed

#To access the first 3 elements
print("The first 3 elements:")
print(l[0:3]) #['Python', 'Java', 'C']
print(l[:3])  #['Python', 'Java', 'C']

#To access the elements alternately
print("\nThe elements alternately:")
#Here, the step is 2
print(l[0::2]) #['Python', 'C', 'JavaScript']

#To print the list in reverse order
print("\nThe elements in reverse order:")
print(l[::-1]) #['JavaScript', 'C#', 'C', 'Java', 'Python']

#To access the last 3 elements
print("\nThe last 3 elements:")
print(l[-3:]) #['C', 'C#', 'JavaScript']

#To access all elements except the last 3 elements
print("\nAll elements except the last 3 elements:")
print(l[:-3]) #['Python', 'Java']

#Here you have to calculate the 'length of the list - 3' ==> 5-3=2 
print(l[:2]) #['Python', 'Java']
```

#### **Inserting Elements**

```python
print("The original list is:",l)
```

```python
#append(element) --> Inserts an element at the end of the list
print("The list after appending C++ is:")
l.append("C++") #['Python', 'Java', 'C', 'C#', 'JavaScript', 'C++']
print(l)
```

```python
#insert(index,element) --> Inserts an element at the given index
print("The list after inserting Kotlin at index 2 is:")
l.insert(2,"Kotlin") #['Python', 'Java', 'Kotlin', 'C', 'C#', 'JavaScript', 'C++']
print(l)
```

```python
#extend(iterable) --> Insert a list at the end of another list
print("The list after inserting [Scala,Swift] at the end is:")
l.extend(["Scala","Swift"]) #['Python', 'Java', 'Kotlin', 'C', 'C#', 'JavaScript', 'C++', 'Scala', 'Swift']
print(l)
```

```python
#Storing the original list back in l
l=l_copy
```

#### **Removing Elements**

```python
print("The original list is:",l)
```

```python
#pop(index) --> Removes the element specified at index i. If no value is given then it removes the last element
print("The list after popping the first element is:")
l.pop(0) #['Java', 'C', 'C#', 'JavaScript']
print(l)
```

```python
#remove(element) --> Removes the first occurence of the element
print("The list after removing Java is:")
l.remove("Java") #['C', 'C#', 'JavaScript']
print(l)
```

```python
#clear() --> Removes all the elements from the list
print("The list after clear is:")
l.clear() #[]
print(l)
```

```python
#Storing the original list back in l
l=['Python', 'Java', 'C', 'C#', 'JavaScript']
```

#### **Sorting Elements**

```python
print("The original list is:",l)
```

```python
#sort(reverse=False) --> Sorts the elements in ascending order. reverse = False is the default argument
print("The list after sorting in ascending order:")
l.sort() #['C', 'C#', 'Java', 'JavaScript', 'Python']
print(l)

print("\nThe list after sorting in descending order:")
l.sort(reverse=True)
print(l) #['Python', 'JavaScript', 'Java', 'C#', 'C']
```

```python
#sort(reverse=False) --> Reverses the list
print("The list before reversing:")
print(l) #['Python', 'JavaScript', 'Java', 'C#', 'C']

print("\nThe list after reversing:")
l.reverse() #['C', 'C#', 'Java', 'JavaScript', 'Python']
print(l)
```

```python
#Storing the original list back in l
l=['Python', 'Java', 'C', 'C#', 'JavaScript']
```

#### **Counting the occurence of an Element**

```python
#count(element) --> Counts all the occurences of a given element
l.append("Python")
print(l)
print("\nThe of times Python appears is:",l.count("Python")) #2
```

#### **Built-in Methods**

```python
print("The original list is:",num_list)

#sum(list) --> The sum of all the elements in the list
print("\nThe sum of all the elements in the list is:",sum(num_list)) #24

#min(list) --> The minimum value in the list
print("\nThe least element in the list is:",min(num_list)) #1

#max(list) --> The maximum value in the list
print("\nThe greatest element in the list is:",max(num_list)) #6

#len(list) --> The length of the list
print("\nThe length of the list is:",len(num_list)) #8

#set(list) --> Removes duplicates by converting the list to a set
#set is converted back to list for displaying
print("\nThe list without duplicates is:",list(set(num_list))) #[1, 2, 3, 4, 5, 6]

#enumerate(list) --> Prints all the elements of the list along with their indices
print("\nThe enumeration of the list is:")
for i,e in enumerate(num_list):
  print(i,e)

#zip(list1,list2) --> Zips the elements of List 1 with List 2
l1=["A","B","C"]
l2=[1,2,3]
print("\nThe zip of",l1,"and",l2,"is:")
for i in zip(l1,l2):
  print(i)
```

### **How to check if an element is in the list?**
'in' and 'not in' can be used for this purpose

```python
print("The original list is:",l)
if "Java" in l:
  print("\nJava is present")
else:
   print("\nJava is not present")

if "HTML" not in l:
   print("HTML is not present")
else:
   print("HTML is present")
```

### **List Comprehensions**
Provides a consice way to create lists from loops

```python
#Suppose we want to double every element present in num_list and store that in a new list
print("The original list is:",num_list)
d=[]
#Without List Comprehensions
for e in num_list:
  d.append(e*2)
print(d)

#Using List Comprehensions
d=[e*2 for e in num_list]
print(d)

#Answer: [6, 8, 4, 4, 2, 2, 10, 12]
```

### **2D Lists**
These are similar to a 2D array. Each element has to be indexed using two values

```python
two_list=[[1,2,3],[4,5,6]]
print("The list is:")
print(two_list)

#The element present in the first row and first column --> (0,0)
print("\nThe element present in the first row and first column is:",two_list[0][0]) #1

#The element present in the second row and third column --> (1,2)
print("The element present in the second row and third column is:",two_list[1][2]) #6
```
