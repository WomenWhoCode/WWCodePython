In Python, anyonymous functions are declared with the *lambda* keyword, hence the term Lambda functions.

## Declaring Lambda Functions

Lambda functions have a pretty simple syntax, they're declared as follows:

```python
lambda a : a * a
```
- ***lambda*** is the keyword to define the function

- **_a_** is the argument for the function

- Everything after the ***:*** is the expression that will be evaluated


Let's take a look some examples below:

````python
# Declaring a Lambda function

# Example 1: function to square a number
square = lambda x: x*x

print(f' Square of 5:  {square(5)}')

print(f' Square of 6:  {square(6)}')
````

````python
# Example 2: function to double a number

doub = lambda x: x*2

print(f' Double of 2:  {doub(2)}')

print(f' Double of 8:  {doub(8)}')
````

You might be wondering about the extra code on the right side of the ***=*** compared to the syntax I showed you. It looks familiar to a variable assignment, and variables are nothing but names for values.

Hold on...*names*? Isn't the whole point of a Lambda function to be used without a name?

## Arguments for a Lambda Function

Arguments for Lambda functions have all the variation as for a normal function, and can accept all types of arguments as well. Here are a few examples of each:

<a name= '21'></a>
### Single Argument

````python
# This is the standard type of argument that has only one parameter and all operations are performed only on it

identity = lambda a: a

square = lambda a: a*a

increment = lambda a: a+1

decrement = lambda a: a-1

var = 10

print(f'''Examples of Single Argument Lambda Functions:\n{var} = {identity(var)}
{var}^2 = {square(var)}\n{var}+1 = {increment(var)}\n{var}+1 = {decrement(var)}''')
````

<a name= '22'></a>
### Multiple Arguments

````python
# Lambda functions can accept multiple arguments as well, surrounded by a comma(,)

power = lambda num, exp: num**exp

add_int = lambda a, b: a+b

a, b = 8, 3

print(f'Examples of Multiple Argument Lambda Functions:\n{a}^{b} = {power(a, b)}\n{a}+{b} = {add_int(a, b)} ')
````

<a name= '23'></a>
###Default Arguments

````python
# You can set default values for the arguments as well

say_name = lambda greet, name="John Doe": f'{greet} {name}'

power = lambda num, exp=2: num**exp

var = 9

print(f'Examples of Default Argument Lambda Functions:\n{say_name("Hello")}\n{var}^2 = {power(var)}')
````

<a name= '24'></a>
###Variable Length Arguments

````python
# And finally, you can even have variable length arguments

sum_ = lambda *nums: sum(nums)

avg_ = lambda *nums: sum(nums)/len(nums)

print(f'Examples of Variable Length Argument Lambda Functions:\nSum of 1, 2, 3 = {sum_(1, 2, 3)}\nAverage of 1, 2, 3 = {avg_(1, 2, 3)}')
````

## Difference Between Lambda and Defined Functions

The major differences between defined and Lambda functions lies in their syntax, usages and limitations. Let's take a look at them:

<a name= '31'></a>
### Syntax

````python
# Defined function

def cube_int(a):
    return a**3

cube_int_lambda = lambda a: a**3

print(f'Cube of 5 using a defined function: {cube_int(5)}')

print(f'Cube of 5 using a Lambda function: {cube_int_lambda(5)}')
````

<a name= '32'></a>
###Uses

````python
# Defined functions are declared for reusability and are of a more 'permanent' stature

def factorial(n):
    if n == 1:
        return 1
    else:
        return n*factorial(n-1)

a, b = 3, 5

#....

fact1  = factorial(a)

#.....

fact2 = factorial(b)

#....

print(f'Factorial of {a}, {b}, respectively: {fact1}, {fact2}')
````

As you saw above, the *factorial* function was defined once at the beginning and was referenced multiple times throughout the program (the *#....* represents other code as part of a bigger program/aplication).

In this sense, defined functions are more permanent and are declared with reusability in mind

````python
# Whereas Lambda functions are more of a throwaway function and are of a more 'temporary' stature

a, b, c, d, e, f, g = 7, 8, 6, 21, 10, 3, 2

#....

add_mult = (lambda a, b: a+b)(a, b)*c

#....

power_ = (lambda num, exp: num**exp)(e, g)

#....

div_nums = (lambda a, b: d/f)

#....
````

The above situation shows one of the more conventional uses of the Lambda function talked about earlier in this notebook: they serve as more of a throw-away function, written as and when required without having to take up the extra space normal functions take up. 


The way they're used here is also known as *immediate invokation*, which is when a function can be directly used without having to be called outside its definition, similar to the usage of a regular Python function

## Uses of Lambda Function


I've been talking up the conventional use of Lambda functions since the first section, and the wait is over. Before we get into practical examples, let's make clear the use-cases for Lambda funtions (some of which you've already seen in this notebook):

- **Provide readability for simple functions**: Though using Lambda functions with a name isn't conventional, it's still a way to use them. This would be when you have a simple function but don't want to deal with the syntax of a normal function, and this improves readability as well

- **With built-in functions**: Functions like filter(), map(), etc, require a function as input. A simple Lamdba function can be passed in instead of defining a normal function

- **As filters for Lists**: Lambda functions can be used in List Comprehensions, a shorthand way of creating lists with a certain condition

- **For immediate invokation**: Lambda functions can be immediately invoked, instead of having to be seprately called after their definition

With the uses out of the way, let's dive into some examples!



<a name= '41'></a>
### Use With Lists

````python
# Using Lambda functions to create lists

# Create a list with multiples of 2

mult_2 = [(lambda x : x*2)(x) for x in range(1, 10)]

print(f"List of elements that are multiples of 2: {mult_2}")
````

Functions like *filter, map and reduce* expect a ***function*** and a ***list*** as arguments. They are what's known as **higher-order** functions. The full extent of these functions is outside the scope of this notebook, but for our purposes it's best to think fo them as functions that use other smaller functions

<a name= '42'></a>
### Use With Filter

The filter function returns elements in the list that only match the condition mentioned in the function

````python
# Using Lambda to filter out elements in a lists

# Filter out odd elements in a list

nums = [(lambda x: x)(x) for x in range(1, 20)]

odd = list(filter(lambda x: x%2 != 0, nums)) # Suntax: filter(func, list)

# The extra list wrap-around is to convert the output of the filter into a list of human-readable elements

print(f'Example 1:\nList of original elements: {nums} | List of odd elements: {odd}\n\n')

# Filter our elements greater than a certain value

thresh = 20

list_ = [1, 3, 19, 60, 39, 23, 3, 5, 67, 24, 11]

filter_list = list(filter(lambda x: x>thresh, list_))

print(f'Example 2:\nList of original elements: {list_} | List of elements greater than threshold({thresh}): {filter_list}\n\n')

# Find all elements divisible by 3

div_2 = list(filter(lambda x: x%3 == 0, nums))

print(f'Example 3:\nList of original elements: {nums} | List of elements divisible by 3: {div_2}\n\n')

# Find all elements that contain the letter i

strings = ["Grape", "Mango", "Pie", "Kiwi", "Guava"]

cont_i = list(filter(lambda x: "i" in x, strings))

print(f'Example 4:\nList of original elements: {strings} | List of elements containng i: {cont_i}\n\n')

# Find all elements that are palindromes

words = ["nun", "racecar", "frame", "milk", "hannah"]

palind = list(filter(lambda x: x == ''.join(reversed(x)), words))

print(f'Example 5:\nList of original elements: {words} | List of palindromes: {palind}\n\n')

# Find all elements longer than 3 letters

words2 = ["cats", "an", "ant", "by", "banana"]

len_filter = list(filter(lambda x: len(x) >= 3, words2))

print(f'Example 6:\nList of original elements: {words2} | List of words longer than 3 letters: {len_filter}')
````

### Use With Map

The map functions applies a single operation defined in the function on all elements in the list provided

````python
# Squaring every element in a list

nums = [(lambda x: x)(x) for x in range(1, 10)]

squares = list(map(lambda x: x**2, nums))

print(f'Example 1:\nList of original elements: {nums} | List of squared elements: {squares}\n\n')

# Reversing each string in a list

strings = ["grape", "mango", "pie", "kiwi", "guava"]

rev = list(map(lambda x: x[::-1], strings))

print(f'Example 2:\nList of original elements: {strings} | List of reversed elements: {rev}')
````

### Use With Reduce


The reduce function performs the operation in the function repeatedly over the list in pairs to produce a reduced result

````python
# Sum over all elements

from functools import reduce

nums = [5, 8, 50, 100, 10, 20]

sum_ = reduce((lambda x, y: x + y), nums) #results of previous two elements are added to the next element and this goes on till the end of the list like (((((5+8)+50)+100)+10)+20)

print(f'Example 1:\nList of original elements: {nums} | Sum of elements: {sum_}\n\n')

# Find maximum element in list

max_ = reduce(lambda a,b : a if a > b else b,nums)

print(f'Example 2:\nList of original elements: {nums} | Maximum of elements: {max_}')
````
