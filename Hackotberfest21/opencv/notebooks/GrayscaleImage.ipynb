```python
import cv2
import matplotlib.pyplot as plt
```

4. Grayscaling of images

  cv2.cvtColor(image,colorcode)

Parameter:

- image: source image read
- colorcode: color space conversion code. for grayscale colorcode is cv2.COLOR_BGR2GRAY



```python
image = cv2.imread('puppy.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY )
plt.imshow(gray_image,cmap='gray')  
plt.show() 
```

```python
gray1_image = cv2.imread('puppy.jpg',0)
plt.imshow(gray1_image,cmap='gray')  
plt.show() 
```
